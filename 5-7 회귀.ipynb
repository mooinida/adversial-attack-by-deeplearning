```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
w0=2.3
w1=1
num_data=100
noise=np.random.normal(0,3,num_data)

X=np.linspace(0,5,num_data)
y=w0+w1*(X**2)+noise

plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('y')
plt.margins(0.2)
plt.show()
```

```python
print(noise)
```

```python
X=X.reshape(-1,1)#학습을 위한 차원 변환 진행

#MSE 최소화 결정 트리 생성
dt_mse_reg=DecisionTreeRegressor(max_depth=5)
dt_mse_reg.fit(X,y)
```

```python
from sklearn.tree import plot_tree
# 트리 시각화
plt.figure(figsize=(24, 12))
plot_tree(dt_mse_reg, filled=True,
          feature_names=['Feature 1', 'Feature 2'])
plt.title("Decision Tree using MSE")
plt.show()
```

```python
X_test=np.arange(0.0,5.0,0.01)[:,np.newaxis]
y_pred=dt_mse_reg.predict(X_test)

plt.figure(figsize=(10,6))
plt.scatter(X,y,
            s=20,edgecolor="black",
            c="darkorange",label="data")
plt.plot(X_test,y_pred,
         color="cornflowerblue",label="prediction",linewidth=2)

plt.xlabel("data")
plt.ylabel("target")
plt.title("DD")
plt.legend()
plt.show()
```
